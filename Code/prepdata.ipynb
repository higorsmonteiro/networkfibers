```python
import os
import numpy as np
import pandas as pd
from collections import defaultdict
```

```python
filepath = "../Data/Ecoli/network_tf_gene.txt"
GeneData = pd.read_csv(filepath, sep="\t", comment="#", names=['TFactor', 'RegGene', 'RegType', "Dummy1", "Dummy2", "Dummy3"])
GeneData = GeneData.drop(['Dummy1', 'Dummy2', 'Dummy3'], axis=1)
GeneData['TFactor'] = GeneData['TFactor'].str.lower()
GeneData['RegGene'] = GeneData['RegGene'].str.lower()
GeneData
```

```python
rename = defaultdict(list)
nlines = GeneData.count()[0]

count = 0
for k in range(nlines):
    gene_name1 = GeneData['TFactor'].loc[k]
    gene_name2 = GeneData['RegGene'].loc[k]
    if len(rename[gene_name1])==0:
        rename[gene_name1].append(count)
        count += 1
    else: pass
    if len(rename[gene_name2])==0:
        rename[gene_name2].append(count)
        count += 1

NumberedGene = GeneData.copy()
    
for k in range(nlines):
    gene_name1 = GeneData['TFactor'].loc[k]
    gene_name2 = GeneData['RegGene'].loc[k]
    NumberedGene['TFactor'].loc[k] = rename[gene_name1][0]
    NumberedGene['RegGene'].loc[k] = rename[gene_name2][0]
    
NumberedGene
```

```python
NumberedGene.to_csv("../Data/Ecoli/ecoliedgelist.dat", sep="\t", header=False, index=False, columns=['TFactor', 'RegGene', 'RegType'])
```

```python
filepath1 = "../Data/Ecoli/Ecoli.txt"
GeneData1 = pd.read_csv(filepath1, sep=" ", comment="#", names=['TFactor', 'RegGene', 'RegType'])
nlines1 = GeneData1.count()[0]
print(nlines1)

GeneData1
```

```python
rename1 = defaultdict(list)

count = 0
for k in range(nlines1):
    gene_name1 = GeneData1['TFactor'].loc[k]
    gene_name2 = GeneData1['RegGene'].loc[k]
    if len(rename1[gene_name1])==0:
        rename1[gene_name1].append(count)
        count += 1
    else: pass
    if len(rename1[gene_name2])==0:
        rename1[gene_name2].append(count)
        count += 1
    else: pass

NumberedGene1 = GeneData1.copy()
    
for k in range(nlines1):
    gene_name1 = GeneData1['TFactor'].loc[k]
    gene_name2 = GeneData1['RegGene'].loc[k]
    reg = GeneData1['RegType'].loc[k]
    NumberedGene1['TFactor'].loc[k] = rename1[gene_name1][0]
    NumberedGene1['RegGene'].loc[k] = rename1[gene_name2][0]
    if reg=='positive': NumberedGene1['RegType'].loc[k] = "+"
    elif reg=='negative': NumberedGene1['RegType'].loc[k] = "-"
    elif reg=='dual': NumberedGene1['RegType'].loc[k] = "+-"
    else: NumberedGene1['RegType'].loc[k] = "n"
    
NumberedGene1
```

```python
NumberedGene1.to_csv("../Data/ecoliedgelist.dat", sep="\t", header=False, index=False, columns=['TFactor', 'RegGene', 'RegType'])
np.savetxt("../Data/ecoliN.dat", [count], fmt="%d")
```
